```python
import matplotlib.pyplot as plt
```

```python
r_sun = 1
w_sun = 1
```

```python
r_planet = 2
w_planet = 0.5
phase = 0.3
```

```python
import numpy as np

theta = np.linspace(0, 2*np.pi, 200)
```

```python
x_sun = r_sun*np.cos(w_sun*theta)
y_sun = r_sun*np.sin(w_sun*theta)
```

```python
x_planet = r_planet*np.cos(w_planet*theta + phase)
y_planet = r_planet*np.sin(w_planet*theta + phase)
```

```python
plt.plot(x_sun, y_sun, 'y', linewidth=3)
plt.plot(x_planet, y_planet, 'b', linewidth=1)
plt.axis('equal')
plt.show()
```
